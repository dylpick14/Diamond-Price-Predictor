## Seperate Data into Training & Testing Sets
Select the:
- percentage of data to be used for classic test/validation split training
- number of folds for cross-validation

```python
# Percentage (0-1.0 corresponds to 0% to 100%) of dataset
percent_training_data= .8
precent_validation_data = 1-percent_training_data
num_cv_folds = 5

```

Now separate the data into training and validation sets by setting the percentage of data to be used for training:

```python
# Choose the first (percent_training_data)% examples for training
num_total_examples = len(df_diamond)
num_training_examples = math.ceil(num_total_examples * percent_training_data)
num_validation_examples = num_total_examples - num_training_examples

# Get all examples (useful later on...)
df_features_all = get_features_dataframe(df_diamond.head(num_total_examples))
df_targets_all = get_targets_dataframe(df_diamond.head(num_total_examples))

# Choose the first (percent_training_data)% for training examples
df_features_training = get_features_dataframe(df_diamond.head(num_training_examples))
df_targets_training = get_targets_dataframe(df_diamond.head(num_training_examples))

# Choose the last (1-percent_training_data)% for validation examples
df_features_validation = get_features_dataframe(df_diamond.tail(num_validation_examples))
df_targets_validation = get_targets_dataframe(df_diamond.tail(num_validation_examples))
```

## Train Data
The following code fits several classifiers to the training data:

```python
#Train/fit Ridge regression model
from sklearn.linear_model import Ridge
ridgeReg = Ridge(alpha = 1.0)
ridgeReg.fit(df_features_training_normalized, df_targets_training.to_numpy().ravel())
lst_regmodels.append(ridgeReg)
print(ridgeReg, end="\n\n\t")

#Train/fit Linear Support Vector Regression model
from sklearn.svm import LinearSVR
lsvr =LinearSVR(random_state=0, tol=1e-5, max_iter = 100000)
lsvr.fit(df_features_training_normalized, df_targets_training.to_numpy().ravel())
lst_regmodels.append(lsvr)
print(lsvr, end="\n\n\t")



```
